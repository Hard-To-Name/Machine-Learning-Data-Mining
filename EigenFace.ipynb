```python
import numpy as np
import mltools as ml
from scipy import linalg
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = np.genfromtxt("data/iris.txt", delimiter = None)
X, Y = iris[:, 0:2], iris[:, -1]
```

## Part (1)

```python
f1, f2 = X[:, 0], X[:, 1]
plt.scatter(f1, f2)
```

## Part (2)

```python
best = [np.array([]), np.array([]), np.array([])]
best_scores = [1e9, 1e9, 1e9]
```

###  1st Initialization

```python
fig, axs = plt.subplots(nrows = 1, ncols = 3)
k_vals = [2, 5, 20]
for k in range(len(k_vals)):
    z, c, sumd = ml.cluster.kmeans(X, k_vals[k])
    axs[k].scatter(f1, f2, c = z)
    axs[k].set_title("K = %d" % k_vals[k])
    print("When K = %d, sum of squared euclidean distances =" 
          % k_vals[k], sumd)
    if sumd < best_scores[k]:
        best_scores[k] = sumd
        best[k] = np.array([z, c, sumd])
plt.tight_layout()
```

## Part (3)

```python
fig, axs = plt.subplots(nrows = 1, ncols = 3)
# k_vals = [2, 5, 20] as defined in part (2)
for k in range(len(k_vals)):
    z, join = ml.cluster.agglomerative(X, k_vals[k])
    axs[k].scatter(f1, f2, c = z)
    axs[k].set_title("K = %d" % k_vals[k])
plt.tight_layout()
```

## Part (4)

**Similarities:** Both clustering methods continuously adjust clusters base on the shortest paths. While K-means is like Dijkstra (not exactly but both keep updating distances/costs) with K starting point, the agglomerative clustering is like Kruskal. They both serve the clustering purpose.  
  
**Differences:** Each execution of the agglomerative clustering (with "join" not specified) returns the same result; whereas each execution of K-means clustering may return different results according to different initialized centers.

# Problem 2: EigenFaces

```python
X = np.genfromtxt("data/faces.txt", delimiter = None)
```

```python
plt.figure()
img = np.reshape(X[0, :],(24, 24))
plt.imshow(img.T , cmap = "gray")
```

```python
plt.figure()
img = np.reshape(X[3, :],(24, 24))
plt.imshow(img.T , cmap = "gray")
```

## Part (1)

```python
m = np.mean(X, axis = 0)
X0 = X - m
```

```python
img = np.reshape(m,(24, 24))
plt.imshow(img.T , cmap = "gray")
```

## Part (2)

```python
U, s, Vh = linalg.svd(X0, full_matrices = False)
```

```python
W = U.dot(np.diag(s))
```

```python
print("Shape of W:", W.shape)
print("Shape of Vh:", Vh.shape)
```

## Part (3)

```python
MSE_vals = []

for k in range(1, 11):
    X0_hat = np.dot(W[:, :k], Vh[:k, :])
    MSE_vals.append(np.mean((X0 - X0_hat) ** 2))
    
plt.plot(MSE_vals)
plt.xticks(range(1, 11))
plt.xlabel("K")
plt.ylabel("MSE")
```

## Part (4)

```python
alpha = 2 * np.median(np.abs(W), axis = 0)
```

```python
fig, axs = plt.subplots(nrows = 1, ncols = 2)

first_minus = m - alpha[0] * Vh[0, :]
axs[1].imshow(np.reshape(first_minus, (24, 24)).T, cmap = 'gray')
axs[1].set_title("mu - alpha * V[j, :]")

first_plus = m + alpha[0] * Vh[0, :]
axs[0].imshow(np.reshape(first_plus, (24, 24)).T, cmap = 'gray')
axs[0].set_title("mu + alpha * V[j, :]")
```

```python
fig, axs = plt.subplots(nrows = 1, ncols = 2)

second_minus = m - alpha[1] * Vh[1, :]
axs[1].imshow(np.reshape(second_minus, (24, 24)).T, cmap = 'gray')
axs[1].set_title("mu - alpha * V[j, :]")

second_plus = m + alpha[1] * Vh[1, :]
axs[0].imshow(np.reshape(second_plus, (24, 24)).T, cmap = 'gray')
axs[0].set_title("mu + alpha * V[j, :]")
```

```python
fig, axs = plt.subplots(nrows = 1, ncols = 2)

third_minus = m - alpha[2] * Vh[2, :]
axs[1].imshow(np.reshape(third_minus, (24, 24)).T, cmap = 'gray')
axs[1].set_title("mu - alpha * V[j, :]")

third_plus = m + alpha[2] * Vh[2, :]
axs[0].imshow(np.reshape(third_plus, (24, 24)).T, cmap = 'gray')
axs[0].set_title("mu + alpha * V[j, :]")
```
